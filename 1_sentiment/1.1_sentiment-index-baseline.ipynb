The script processes linguistic data by generating samples that combine natural sentences and synthetic variations (both positive and negative) across predefined time intervals for specified target terms. It handles sampling without replacement within each batch to ensure diversity and integrity, and saves the mixed samples into separate directories based on the variation type, applying different injection ratios to simulate varying levels of synthetic data integration for linguistic analysis.
